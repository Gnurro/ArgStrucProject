```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate

# Read Excel file into a DataFrame
df = pd.read_excel('survey_results.xlsx')

df = df.drop(columns=['Timestep'], errors='ignore')

# Display the DataFrame
#df

```

```python

#df.info()

#  summary statistics of numerical columns
#df.describe()

#df.head()

# Calculate the percentage of participants who chose each option for each text question
percentages = {}
for column in df.columns:
    if column.startswith('Text'): 
        counts = df[column].value_counts(normalize=True) * 100
        percentages[column] = counts

```

```python


for column, counts in percentages.items():
    plt.figure(figsize=(6, 4))
    counts.plot(kind='bar', color='skyblue')
    plt.title(f"Percentages for {column} ")
    plt.xlabel ("Options") 
    plt.ylabel("Percentage of Participants")
    plt.xticks(rotation=45)
    plt.xticks([])
    plt.tight_layout()
    plt.show()
```

```python
for column, counts in percentages.items():
    print(f"Percentages for {column}:")
    print("Percentage\tOption")
    for option, percentage in counts.items():
        print(f"{percentage:.2f}%\t\t{option}")
    print()
```

```python


for column, counts in percentages.items():
    print(f"Percentages for {column}:")
    table_data = [(f"{percentage:.2f}%", option) for option, percentage in counts.items()]
    print(tabulate(table_data, headers=['Percentage', 'Option'], tablefmt='grid'))
    print()
```
